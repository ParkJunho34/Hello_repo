# 수집 데이터 정의

- 접속 사이트 : https://search.daum.net/search?w=tot&q=
- 수집 대상 : '핀테크' 키워드 조회결과
- 수집건수 : 가능한 많이
- 저장 위치 : /contents/data

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

search_keyword = '핀테크'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36',}

url = f'https://search.daum.net/search?w=tot&q={search_keyword}'
res = requests.get(url, headers=headers)
soup = BeautifulSoup(res.content.decode('utf-8'), 'html.parser')
```

```python
soup
```

```python
# 수집대상 정보 식별
total = soup.find_all('strong', 'tit-g clamp-g')
total
```

```python
total[0]
```

```python
# 식별된 정보 수집
search_result = []
for i in total:
    temp = []
    temp.append(i.text)  # 제목
    search_result.append(temp)
```

```python
pd.DataFrame(search_result, columns=['title'])
```

```python
# 저장
search_result = pd.DataFrame(search_result, columns=['title'])
search_result.to_csv(f'news_핀테크.csv')
```

# 정제 : 결측치 확인

```python
# 작업 대상 불러오기
search_result = pd.read_csv('news_핀테크.csv')
search_result.head()
```

```python
# 컬럼별 결측치 확인
search_result.isnull().sum()
```

```python
# 컬럼명이 이쁘지 않아요. 컬럼명을 변경합니다.
search_result.columns = ['index', 'title']
search_result.head()
```

```python
# 컬럼명의 index는 데이터프레임의 index와 동일합니다. 삭제해도 되겠습니다.
search_result = search_result.drop('index', axis=1)#인덱스삭제
search_result.head()
```

```python
# 컬럼별 결측치 다시 확인
search_result.isnull().sum()
```

# 정제 : 삭제(deletion) 방법

```python
# title 컬럼의 결측치 12개를 삭제합니다.
search_result = search_result.dropna()

# 삭제결과를 확인합니다.
search_result.isnull().sum()
```

```python
# 전체 데이터를 확인합니다.
search_result
```

# 정제 : 대체(imputation)

## 단순 대체 (최빈값)

```python
# 평균(mean), 중앙값(median), 최빈값(mode)으로 대체
# 수치형 변수에는 평균/중앙값, 범주형 변수에는 최빈값을 사용
```

```python
# 그렇다면 우리는 ? 범주형 변수이니....
```

```python
# 작업 대상 불러오기
search_result = pd.read_csv('news_핀테크.csv')
search_result.head()
```

```python
# 결측치 확인
search_result.isnull().sum()
```

```python
# 최빈값 구하기
mode_value = search_result['title'].mode()
mode_value
```

```python
# 최빈값으로 대체
search_result['title'] = search_result['title'].fillna(mode_value[0])
```

```python
# 확인
search_result
```

```python
# 저장
# 반드시 기존 파일과 다른 파일로 저장
search_result.to_csv(f'news_핀테크_결측치완료.csv', index=False)
```
