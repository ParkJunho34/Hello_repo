# 보여지는 정보

```python
# 패키지 로드
import requests
from bs4 import BeautifulSoup

import os
import json
import pandas as pd
import numpy as np
```

```python
# 웹 사이트 접속 및 데이터 수신
res = requests.get('https://v.daum.net/v/20231022120603705')
```

```python
# 서버 접속 및 데이터 수신 결과코드 (200 - 정상, 그외 HTTP 통신결과 코드에 정의)
res.status_code
```

```python
# 수신 데이터 Header 정보
res.headers
```

```python
# 수신 데이터 Body 정보 (일반적으로 HTML)
web_data = res.content.decode('utf-8')
print(web_data)
```

```python
# 수신 데이터 구조 정비
soup = BeautifulSoup(web_data, 'html.parser')
```

```python
# 기사제목 및 부가정보 수집
# <h3 class="tit_view" data-translation="true">탕후루 대신 바나나, 이걸 추천합니다</h3>
data = soup.find_all('h3', 'tit_view')

print('기사 제목 : ', data[0].text)
print('부가정보 : ', data[0].attrs)
```

# 수집 데이터 정의

- 접속 사이트 : https://search.daum.net/search?w=tot&q=
- 수집 대상 : '핀테크' 키워드 조회결과
- 수집건수 : 가능한 많이
- 저장 위치 : /contents/data

```python
import requests # 웹 스크래핑을 하기 위하여 웹사이트에 접속하는 기능이 있는 패키지
from bs4 import BeautifulSoup # 웹사이트 접속결과 수신된 데이터를 클랜징 파싱 하는 패키지

search_keyword = '핀테크' # 검색 키워드
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36',} # 클라이언트 접속 정보

url = f'https://search.daum.net/search?w=tot&q={search_keyword}' # 검색 키워드를 포함하는 저네 웹 사이트 접속 정보
res = requests.get(url, headers=headers) # 웹 서버에 get 방식으로 데이터를 요청(= 웹서버에 get 방식으로 접속)
soup = BeautifulSoup(res.content.decode('utf-8'), 'html.parser') # 웹 서버에서 수신한 데이터를 클린징, 정제하는 과정
```

```python
res.content # HTML, CSS, JavaScript.....
```

```python
soup
```

```python
# 수집대상 정보 식별
total = soup.find_all('strong', 'tit-g clamp-g')
total
```

```python
# 식별된 정보 수집
search_result = []
for i in total:
    temp = []
    temp.append(i.text)  # 제목
    search_result.append(temp)
```

```python
pd.DataFrame(search_result, columns=['title'])
```

```python
# 저장
search_result = pd.DataFrame(search_result, columns=['title'])
search_result.to_csv(f'news_핀테크.csv')
```
